```python
global_cases = covid_df.groupby('date')['new_cases'].sum().reset_index()

global_cases.head()
```

```python
latest_data = covid_df.sort_values('date').groupby('location').last().reset_index()

latest_data.head()
```

```python
top10 = latest_data.nlargest(10, 'total_cases')

top10[['location', 'total_cases']]
```

```python
fig = px.bar(
    top10,
    x='location',
    y='total_cases',
    title='Top 10 Countries by Total COVID-19 Cases',
    color='total_cases'
)

fig.show()
```

```python
country_data = covid_df[covid_df['continent'].notna()]
```

```python
latest_country_data = (
    country_data.sort_values('date')
    .groupby('location')
    .last()
    .reset_index()
)
```

```python
import plotly.express as px

fig = px.bar(
    top10,
    x='location',
    y='total_cases',
    color='total_cases',
    text='total_cases',
    title='Top 10 Countries by Total COVID-19 Cases'
)

fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')

fig.update_layout(
    xaxis_title='Country',
    yaxis_title='Total Cases',
    template='plotly_white'
)

fig.show()
```

```python
top10_deaths = latest_country_data.nlargest(10, 'total_deaths')

top10_deaths[['location', 'total_deaths']]
```

```python
import plotly.express as px

fig = px.bar(
    top10_deaths,
    x='location',
    y='total_deaths',
    color='total_deaths',
    text='total_deaths',
    title='Top 10 Countries by Total Deaths'
)

fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')

fig.update_layout(
    template='plotly_white',
    xaxis_title='Country',
    yaxis_title='Total Deaths'
)

fig.show()
```

```python
latest_country_data['vaccination_rate'] = (
    latest_country_data['people_fully_vaccinated']
    / latest_country_data['population']
) * 100
```

```python
latest_country_data[['location', 'vaccination_rate']].head()
```

```python
top10_vaccinated = latest_country_data.nlargest(
    10,
    'vaccination_rate'
)

top10_vaccinated[['location', 'vaccination_rate']]
```

```python
fig = px.bar(
    top10_vaccinated,
    x='location',
    y='vaccination_rate',
    color='vaccination_rate',
    text='vaccination_rate',
    title='Top 10 Countries by Vaccination Rate (%)'
)

fig.update_traces(
    texttemplate='%{text:.2f}%',
    textposition='outside'
)

fig.update_layout(
    template='plotly_white',
    xaxis_title='Country',
    yaxis_title='Vaccination Rate (%)'
)

fig.show()
```

```python
latest_country_data['death_rate'] = (
    latest_country_data['total_deaths']
    / latest_country_data['total_cases']
) * 100
```

```python
top10_death_rate = latest_country_data.nlargest(
    10,
    'death_rate'
)

top10_death_rate[['location', 'death_rate']]
```

```python
fig = px.bar(
    top10_death_rate,
    x='location',
    y='death_rate',
    color='death_rate',
    text='death_rate',
    title='Top 10 Countries by Death Rate (%)'
)

fig.update_traces(
    texttemplate='%{text:.2f}%',
    textposition='outside'
)

fig.update_layout(
    template='plotly_white',
    xaxis_title='Country',
    yaxis_title='Death Rate (%)'
)

fig.show()
```
